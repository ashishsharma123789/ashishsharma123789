```python
import random

# Function to get computer's choice
def get_computer_choice():
    options = ["rock", "paper", "scissors"]
    return random.choice(options)

# Function to determine the winner of a round
def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "tie"
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        return "user"
    else:
        return "computer"

# Main game function
def play_game():
    print("Welcome to the Rock-Paper-Scissors Game!")
    print("Instructions:")
    print("- Choose 'rock', 'paper', or 'scissors' to play a round.")
    print("- Type 'q' anytime to quit the game.")
    print("- The game will display the scores after each round.")
    print("- Good luck and have fun!\n")

    user_score = 0
    computer_score = 0
    round_number = 1

    while True:
        print(f"\n--- Round {round_number} ---")
        print("Enter your choice:")
        print("1. Rock")
        print("2. Paper")
        print("3. Scissors")
        print("Type 'q' to quit")

        # Get user's choice
        user_choice = input("Your choice (rock, paper, scissors, or 'q' to quit): ").lower()

        # Handle quit option
        if user_choice == 'q':
            print("\nGame Over! Thank you for playing!")
            print(f"Final Score - You: {user_score}, Computer: {computer_score}")
            break

        # Validate user input
        if user_choice not in ["rock", "paper", "scissors"]:
            print("Invalid choice! Please enter 'rock', 'paper', or 'scissors'.")
            continue

        # Get computer's choice
        computer_choice = get_computer_choice()
        print(f"Computer chose: {computer_choice}")

        # Determine the winner
        winner = determine_winner(user_choice, computer_choice)

        # Display round result and update score
        if winner == "user":
            user_score += 1
            print("üéâ You win this round!")
        elif winner == "computer":
            computer_score += 1
            print("üíª Computer wins this round!")
        else:
            print("ü§ù It's a tie!")

        # Display the current score
        print(f"Scoreboard - You: {user_score}, Computer: {computer_score}")

        # Ask if the user wants to play another round
        play_again = input("\nDo you want to play another round? (y/n): ").lower()
        if play_again != 'y':
            print("\nThank you for playing!")
            print(f"Final Score - You: {user_score}, Computer: {computer_score}")
            break

        # Increment round number for next iteration
        round_number += 1

# Run the game
play_game()
```
